```python
import pandas as pd
```

```python
df = pd.read_csv('dataset.csv')

moda = df['Quantidade'].value_counts().index[0]

df['Quantidade'] = df['Quantidade'].fillna(moda)

df.head()
```

```python
# .mean = media em ingles
# Quero os dados da primeira lista, agrupando pelas colunas que não sao valores

df[['Segmento', 'Regiao', 'Valor_Venda']].groupby(['Segmento', 'Regiao']).mean()
```

```python
# Quero os dados da primeira lista, agrupando pelas colunas que não sao valores, mas trazendo mais indicadores numericos

df[['Segmento', 'Regiao', 'Valor_Venda']].groupby(['Segmento', 'Regiao']).agg(['mean', 'std', 'count'])
```
