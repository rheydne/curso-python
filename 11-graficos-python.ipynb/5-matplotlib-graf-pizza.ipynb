```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
fatias = [7, 2, 2, 13]
atividades = ['dormir', 'comer', 'passear', 'trabalhar']
cores = ['olive', 'lime', 'violet', 'royalblue']
```

```python
plt.pie(fatias, labels=atividades, colors=cores, startangle=90, shadow=True, explode=(0, 0.2, 0, 0), autopct='%1.1f%%')
plt.show()
```
