```python
# Imports
import random
import numpy as np
import pandas as pd
import matplotlib as mat
import matplotlib.pyplot as plt
import warnings
import seaborn as sea
warnings.filterwarnings("ignore")
```

```python
# Carregando um dos datasets que vem com o Seaborn
dados = sea.load_dataset("tips")

dados.head()
```

```python
# O mtodo joinplot cria plot de 2 variveis com grficos bivariados e univariados
sea.jointplot(data = dados, x = "total_bill", y = "tip", kind = 'reg')
```

```python
# O mtodo lmplot() cria plot com dados e modelos de regresso
sea.lmplot(data = dados, x = "total_bill", y = "tip", col = "smoker")
```

```python
# Construindo um dataframe com Pandas
df = pd.DataFrame()

# Alimentando o Dataframe com valores aleatrios
df['idade'] = random.sample(range(20, 100), 30)
df['peso'] = random.sample(range(55, 150), 30)
```
