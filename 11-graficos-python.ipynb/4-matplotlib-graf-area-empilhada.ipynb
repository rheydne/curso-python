```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
dias = [1,2,3,4,5]
dormir = [7,8,6,77,7]
comer = [2,3,4,5,3]
trabalhar = [7,8,7,2,2]
passear = [8,5,7,8,13]
```

```python
plt.stackplot(dias, dormir, comer, trabalhar, passear, colors=['m', 'c', 'r', 'k'])
plt.title("Distribuição de Atividades Diárias")
plt.xlabel("Dias")
plt.ylabel("Horas")
plt.legend(labels=['Dormir', 'Comer', 'Trabalhar', 'Passear'], loc='upper left')
plt.show()
```
